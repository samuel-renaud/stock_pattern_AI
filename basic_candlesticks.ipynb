<H2> This notebook is an intro notebook that allows you to graph a candlestick chart for any stock you are interested in over any time period you specify

```python
ticker = 'QQQ'#<- enter the ticker you are interested in
period = '5mo' #timeframe (1y, 2mo, 200d)

import plotly.graph_objects as go
import yfinance as yf
ticker = yf.Ticker(ticker)
df = ticker.history(period=period)

fig = go.Figure(data=[go.Candlestick(x=df.index,
                                     open=df['Open'],
                                     high=df['High'],
                                     low=df['Low'],
                                     close=df['Close'])],
                )

fig.show()


```
